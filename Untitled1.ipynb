```python
def cir(x,r=1):
    return np.cos(x)*r,np.sin(x)*r

x=np.linspace(-np.pi,np.pi,200)
for m in range(60):
    plt.plot([np.cos(m*np.pi/30.0),np.cos(m*np.pi/30.0)*0.9],[np.sin(m*np.pi/30.0),np.sin(m*np.pi/30.0)*0.9],'k')
for m in range(12):
    plt.plot([np.cos(m*np.pi/6.0),np.cos(m*np.pi/6.0)*0.8],[np.sin(m*np.pi/6.0),np.sin(m*np.pi/6.0
)*0.8],'k')
    plt.text(np.cos(m*np.pi/6.0)*0.65-0.03, np.sin(m*np.pi/6.0)*0.65-0.03, str((-m+14)%12+1),
fontsize=16)
plt.plot(*cir(x),'k')
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
from matplotlib import pyplot as plt

for i in range(3):
    choice_data = iris.data[iris.target==i]
    plt.plot(choice_data[:,0],choice_data[:,1],'o',label=iris.target_names[i])

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend()
plt.show()
```

```python
ass=['sl','sw','pl','pw']
for z in range(4):
    for j in range(4):
        ax=1+z*4+j
        plt.subplot(4,4,ax)
        
        for i in range(3):
            choice_data = iris.data[iris.target==i]
            plt.plot(choice_data[:,z],choice_data[:,j],'o')
        s=ass[z]+'-'+ass[j]
        plt.title(s)
        
plt.tight_layout()        
plt.show()
```
